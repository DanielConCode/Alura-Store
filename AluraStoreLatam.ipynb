### Importación de datos



````python
import pandas as pd

# pandas es una librería de Python que te permite trabajar con estructuras de datos tipo tablas

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# La función read_csv() sirve para leer archivos .csv (Comma Separated Values) y convertirlos automáticamente en un DataFrame de pandas.
# DataFrame → una tabla completa (como una hoja de cálculo).

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
# El método head() se usa para ver las primeras filas de un DataFrame. Esto te muestra las primeras 5 filas por defecto pero También se puede especificar cuántas filas deseamos ver agregandolo como parametro de head().
````

#1. Análisis de facturación



## Ingreso total por cada tienda

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

````python
'''
Para calcular el total de una columna de un DataFrame, utilizamos la sintaxix dataframe['nombre_columna'].sum()
'''

ingreso1 = float(tienda['Precio'].sum())
ingreso2 = float(tienda2['Precio'].sum())
ingreso3 = float(tienda3['Precio'].sum())
ingreso4 = float(tienda4['Precio'].sum())

ingresos = [ingreso1, ingreso2, ingreso3, ingreso4]

for i in range (0, len(ingresos)):
  print(f'El costo de ingreso de la tienda {i+1} es: {ingresos[i]}')
````

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

````python
'''
Para obtener la cantidad de productos vendidos por categoria, podemos usar la funcion .value_conts(), este metodo cuenta la cantidad de veces que aparece cada valor único en una columna de un DataFrame o en una Serie.

La sintaxix es similar a la que utilizamos para sumar, dataframe['nombre_columna'].value_conts()
'''

conteoCategoriaTienda1 = tienda['Categoría del Producto'].value_counts()
conteoCategoriaTienda2 = tienda2['Categoría del Producto'].value_counts()
conteoCategoriaTienda3 = tienda3['Categoría del Producto'].value_counts()
conteoCategoriaTienda4 = tienda4['Categoría del Producto'].value_counts()

print(conteoCategoriaTienda4)
````

````python
'''
Para obtener la categoria mas popular de cada tienda, obtendremos el valor mas repetido, usando .index y .values para crear una lista de tuplas que contendra las categorias mas populares y sus ventas

al obtener el valor con .values, este retornara un int que usa Panda, por lo que lo convertimos a un int normal
'''

categoriaTienda1 = conteoCategoriaTienda1.index[0], int(conteoCategoriaTienda1.values[0])
categoriaTienda2 = conteoCategoriaTienda2.index[0], int(conteoCategoriaTienda2.values[0])
categoriaTienda3 = conteoCategoriaTienda3.index[0], int(conteoCategoriaTienda3.values[0])
categoriaTienda4 = conteoCategoriaTienda4.index[0], int(conteoCategoriaTienda4.values[0])

categoriasPopulares = [categoriaTienda1, categoriaTienda2, categoriaTienda3, categoriaTienda4]

for i in range(0,len(categoriasPopulares)):
  print(f'La categoria mas popular de la tienda {i+1} es {categoriasPopulares[i][0]} con {categoriasPopulares[i][1]} ventas')
````

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

````python
promedio1 = float(tienda['Calificación'].sum() / len(tienda['Calificación']))
promedio2 = float(tienda2['Calificación'].sum() / len(tienda2['Calificación']))
promedio3 = float(tienda3['Calificación'].sum() / len(tienda3['Calificación']))
promedio4 = float(tienda4['Calificación'].sum() / len(tienda4['Calificación']))

promediosSatisfaccion = [promedio1, promedio2, promedio4, promedio4]

for i in range(0,len(promediosSatisfaccion)):
  print(f'La calificacion de los clientes para la tienda {i+1} es {promediosSatisfaccion[i]:.2f}')
````
